```python
import os

files = os.listdir('.')
```

```python
import pandas as pd

full_data = pd.read_csv('data/full_dataset.csv')
```

```python
results = {}

file_prefix = 'llama2'


for file in files:
    if file.startswith(file_prefix) and file.endswith('.txt'):
        print(file)
        with open(file, 'r') as f:
            
            data = f.readlines()
            if 'multitarget' in file:
                pass
            elif False:
                temp = [x.strip() for x in data]
                print(len(temp))
                start = -1
                fn = file.split('.')[0]
                results[fn] = []
                for i in range(len(temp)):
                    if (temp[i].startswith('Yes') or temp[i].startswith('No') or temp[i].startswith('Unclear')):
                        if start >= 0:
                            results[fn].append('\n'.join(temp[start:i]).strip())
                        start = i
    #                 else:
    #                     print(temp[i])
                results[fn].append('\n'.join(temp[start:i]))
            else:
                results[file.split('.')[0]] = [x.strip().split('A: ')[-1] for x in ''.join(data).split('\n------------------******-----------\n')]
```

```python
from collections import Counter
import re
from nltk.tokenize import word_tokenize


hallucinations = 0

predictions = {}
bag_of_words = {}
print(full_data.iloc[0]['text'])
for k, v in results.items():
    if len(full_data) < len(v):
        res = v[-len(full_data):]
    else:
        res = v
        
    print(k,res[0])
    pred = []
    bow = []
    for r in res:
        r = r.split('[\INST]\n')[-1]
        coarse = (2 if r.strip().startswith('Yes') else (1 if r.strip().startswith('Unclear') else (0 if r.strip().startswith('No') else -1)))
        if coarse == -1:
            hallucinations += 1
            if ('Yes' in r.strip()) and not 'Unclear' in r.strip():
                coarse = 2
            elif 'Unclear' in r.strip() and not 'No' in r.strip():
                coarse = 1
            elif 'No' in r.strip():
                coarse = 0
            else:
                coarse = -1
        
        words = set(word_tokenize(r.lower()))
        
#         print(r, words)
        pred.append(coarse)
        bow.append(words)
    predictions[k] = pred
    bag_of_words[k] = bow
#     print(Counter(predictions[k]))
    
#     hallucinations += Counter(predictions[k])[-1]
```
